# Data Preporocessing

```python
df[df.duplicated()]
```

```python
df.isnull().sum()
```

```python
df[df['certificate'].isna()]
```

```python
df['certificate'].unique()
```

```python
df['certificate']= df['certificate'].fillna('Not Rated')
```

```python
df.isnull().sum()
```

```python
df['certificate'].unique()
```

```python
df[df['certificate']=='GP']
```

```python
# We need to change the certification as PG

df['certificate']= np.where(df['certificate']=='GP','PG', df['certificate'])
```

```python
df['certificate'].unique()
```

```python
df[df['certificate']=='Unrated']
```

```python
# It seems both unrated and Not Rated are same as per IMDB.

df['certificate']=np.where(df['certificate']=='Unrated','Not Rated', df['certificate'])
```

```python
df['certificate'].unique()
```

```python
df['duration'].unique()
```

```python
# We need to convert hours and minutes to only in minutes for uniformity.

df['duration']=df['duration'].str.replace("h",'*60').str.replace("m",'*1').str.replace(" ",'+').apply(eval)
```

```python
df['duration'].unique()
```

```python
df.info()
```

```python
df['imbd_rating'].unique()
```

```python
df['genre'].unique()
```

```python
sorted(df['genre'])
```

```python
df['imbd_votes'].unique()
```

```python
# convert IMDB votes to numeric

df['imbd_votes']=df['imbd_votes'].str.replace(",",'').apply(eval)
```

```python
df['imbd_votes'].unique()
```

```python
df.info()
```
